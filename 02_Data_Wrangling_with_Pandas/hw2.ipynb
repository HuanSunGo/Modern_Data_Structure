## Selection of Data

**1. Find the corresponding variable according to the query.**

```python
raw_data['Question'].unique()
```

```python
raw_data['DataValueType'].unique()
```

```python
raw_data['Stratification1'].unique()
```

**2. Remove all columns that are not needed for the analysis.**

```python
new_data=raw_data[(raw_data['Question']=='Poverty') | (raw_data['Question']=='Binge drinking prevalence among adults aged >= 18 years')]
new_data.head(3)
```

```python
new_data_trimmed=new_data[['YearStart','LocationAbbr','LocationDesc','Question','DataValueType','Stratification1'
,'DataValue']]
new_data_trimmed.head(3)
```

**3. Use the selected colums to futher select the data.**

- Getting the binge drinking data.

```python
# Select binge drinking prevalence among adults aged >= 18 years, Crude Prevalence in Percent , overall as well separately for female and males.

binge=new_data_trimmed[(new_data_trimmed['Question']=='Binge drinking prevalence among adults aged >= 18 years')
                &(new_data_trimmed['DataValueType']=='Crude Prevalence')
                &(new_data_trimmed['Stratification1'].isin(['Overall','Female','Male']))]
binge.head(3)
```

```python
# Convert dataset to wide format.

binge_wide=binge.pivot(index=['YearStart','LocationAbbr','LocationDesc'], columns='Stratification1',values='DataValue')
binge_wide.head(3)
```

```python
# Organize, reset and rename the index. 
binge_final=binge_wide.reset_index().rename_axis(None, axis=1)
binge_final.head(3)
```

- Getting the proverty data.

```python
# Select Poverty, Crude Prevalence in Percent, in Overall level.

poverty_data=new_data_trimmed[(new_data_trimmed['Question']=='Poverty')
                &(new_data_trimmed['DataValueType']=='Crude Prevalence')
                &(new_data_trimmed['Stratification1']=='Overall')]
poverty_data.head(3)
```

```python
# Select the revelant data for easier merge later.

poverty_final=poverty_data[['YearStart','LocationAbbr','LocationDesc','DataValue']]
poverty_final.head(3)
```

- Merging two DataFrames together and rename the output.

```python
final_data=pd.merge(binge_final,poverty_final,on=['YearStart','LocationAbbr','LocationDesc'],how='left').rename(
            columns={'YearStart':'year',
              'LocationAbbr':'state_abb',
              'LocationDesc':'state',
              'Female':'binge_female',
              'Male':'binge_male',
              'Overall':'binge_all',
              'DataValue':'poverty'})
final_data.head(3)
```

## Describtio of Data

```python
# Change the data type for the purpose of statistical summary.
final_data[['binge_female','binge_male','binge_all','poverty']]=final_data[['binge_female','binge_male','binge_all','poverty']].astype(float)
final_data.describe()
```

## Analysis of Data

- Produce a table that shows the overall, female, and male binge drinking prevalences across U.S. States in the most recent year of data for the Top 10 binge drinking states.

```python
final_data.head(3)
```

```python
# Determine which year is the most recent year from the data.

final_data['year'].max()
```

```python
# Create a filter for the 2019 binge drinking data.
table=final_data.loc[final_data['year']==2019,['state','state_abb','binge_female','binge_male','binge_all']]

# Change the type of data to float and sort the data by the decending orderï¼Œand chose the top 10 rows.
table[['binge_female','binge_male','binge_all']]=table[['binge_female','binge_male','binge_all']].astype(float)
table.sort_values(by='binge_all',ascending=False).head(10)
```

- Calculate the average annual growth rates (in percent) of overall binge drinking across states for the years the data is available. 

```python
print(final_data['year'].max(),final_data['year'].min())
```

```python
# Transform the dataset to the state-year format.
binge_overall = final_data[['state','year','binge_all']]
binge_year = binge_overall.pivot(index=['state'],columns='year',values='binge_all')
binge_year.head()
```

```python
# Calculate the average annual growht rates accross states.
binge_year.pct_change(axis=1).mean(axis=1).head()
```

- Provide a table of the 5 states with the largest increases and the 5 states with the largest decreases in binge drinking prevalence over the time period.

```python
# Create the dataframe for the annual change.
result=pd.DataFrame(binge_year.pct_change(axis=1).mean(axis=1))
result.columns=['annual_change']

# Select the 5 states with the largest decrease in last_5, and 5 states with the largest increases in top_5.
result['rank']=result.rank(axis=0)
last_5=result.sort_values(by='annual_change',ascending=True).head(5)
top_5=result.sort_values(by='annual_change',ascending=False).head(5)

# Create the visibal table by concating the above two datasets.
pd.concat([last_5,top_5])
```
