### 4.Tortured Truth
#### a) Left join the trial and confession tables. For what share of trials does the database record confessions? Create a results table with the number of all trials, the number of confessions, and the share of trials with confessions recorded.

```python
%%sql
DESCRIBE confession
```

```python
%%sql 
SELECT
COUNT(DISTINCT(t.trialref)) as num_trial,
COUNT(DISTINCT(c.trialref)) as num_conf,
(COUNT(DISTINCT(c.trialref))/COUNT(DISTINCT(t.trialref))) as trials_with_confessions
FROM trial t
LEFT JOIN confession c
ON t.trialref=c.trialref
```

#### b) Only a small number of trials have records of torture. Is there a higher share of confessions among trials with records of torture than trials without such record? Hint: You will need to merge on the confession table.

```python
%%sql 
DESCRIBE torture
```

```python
%%sql

SELECT COUNT(c.trialref)/COUNT(t.trialref) trials_with_confessions
FROM trial t
LEFT JOIN confession c ON t.trialref = c.trialref
LEFT JOIN torture ON c.trialref = torture.trialref
WHERE tortureref IS NOT NULL
```

```python
%%sql

SELECT COUNT(c.trialref)/COUNT(t.trialref) trials_with_confessions
FROM trial t
LEFT JOIN confession c ON t.trialref = c.trialref
LEFT JOIN torture ON c.trialref = torture.trialref
WHERE tortureref IS NULL
```

The result suggests that, the share of trials with the record of torture is 100%, while the share of confessions without torture is 26.96%. Which means the assumption is correct.
